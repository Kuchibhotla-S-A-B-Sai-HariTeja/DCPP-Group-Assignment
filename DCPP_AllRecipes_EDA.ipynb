```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set(style='darkgrid')
```

```python
import json
f = open("all_recipies.json")
data = json.load(f)
```

```python
data[0]
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df_reqd = df[['Title','Carbohydrates (g)','Protein (g)','Fat (g)','Vitamin A (IU)','Calcium (mg)','Iron (mg)']]
df_melt = pd.melt(df_reqd, id_vars = "Title",value_name = "Nutrients")
df_melt.head()
```

```python
df_melt = df_melt.sort_values("Title")
```

```python
sns.factorplot(x = 'Nutrients', y='Title', 
               data = df_melt.head(10), hue = 'variable', kind='bar',);
```

```python
def top5plot(col):    
    plt.figure(figsize=(12,5))
    ax = sns.barplot(data=df.sort_values(col, ascending=False).head(10), y="Title", x=col, palette="Set2")
    for p in ax.patches:
        width = p.get_width()
        plt.text(20+p.get_width(), p.get_y()+0.55*p.get_height(),
                 '{:1.0f}'.format(width),
                 ha='center', va='center')
    plt.title(f"Top 10 Recipes with highest amount of: {col}");
```

```python
top5plot('Calories')
```

```python
top5plot('Sugars (g)')
```

```python
top5plot('Protein (g)')
```

```python
top5plot('Calcium (mg)')
```
